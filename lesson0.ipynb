# 引号的用法

到这里，你已经亲自使用过两次print()函数了。我们来观察一下，你两次写下的print()函数语句有什么不一样的地方呢？
```
print(520)
print('你好')
```

很明显，你会发现，在```print('你好')```里，多了一对单引号。

那么问题又来了：这个单引号是干啥用的？为什么打印文字的时候需要加上引号呢？

这是因为，计算机的脑回路比较特别：只能理解数字，却读不懂文字。因为数字和数学运算是标准化、有固定格式的，而文字却能够千变万化。

如果你直接让计算机```print(你好)```，它会一脸懵并大声报错：“这两个字啥意思，你没有定义清楚，不懂诶……”

````python
# 请直接运行代码，虽然会报错，不过请忽略。并接着往下看
print(你好)
````

这，便是print()函数中引号的用法：当括号内有引号的时候，就相当于告诉计算机——你不需要执行多余的操作，我输入什么，你就打印什么。

此时，计算机会听话地执行你的命令，直接把引号内的内容打印出来。

在print()函数内不仅能使用单引号，还能使用双引号，两者的效果没什么区别，都能让你打印出一行文本。

有些时候，单引号和双引号可能会在括号内同时出现，比如```print("Let's play")```。

这种情况下，需要注意区分清楚哪个引号是属于print()函数结构，哪个引号是属于你要让计算机打印的内容，可别“混搭”了。


说这么多，运行一下代码就懂了。直接运行在下面代码框中的代码，观察运行结果。（运行前先思考一下，在第三个print()函数中，哪个引号才属于print()函数结构）

````python
print('一起玩吧')
#括号内是单引号的情况

print("一起玩吧")
#括号内是双引号的情况

print("Let's play")
#括号内单双引号同时存在的情况
````

很明显，在```print("Let's play")```中，没有被打印出来的引号就属于print()函数结构啦。

不过，在print()函数中，引号里的内容其实也不一定非得是文字，还可以是英文和数字。

但无论你往引号里放啥内容，计算机看到带引号的内容时都会理解为：哦，那我就原样把引号里的内容**复印**一份，显示在终端上吧。

除了**排版文字**之外，**转义字符**还能让计算机**对符号进行区分**。

有时候，我们在打印的过程中，可能会遇到符号重复的问题，尤其是在需要打印英文的时候。比如，下面这个例子：
```
print('I'm not your brother.')
```
这段代码会让计算机报错，因为Python是从左往右匹配单（双）引号的——当它读到一个引号时，会在心里嘀咕：呀，这是不需理解，原样打印的内容。当它再读到下一个引号时，想：好！要打印的内容结束了！

所以，在计算机眼里，上面的`'I'`是要打印的内容，而后面的东西呢，它并不能读懂，于是产生一个报错。

我们来拆解一下这段错误代码：`print('I'm not your brother.')`第一个和第三个单引号属于print()函数的结构，第二个单引号则是纯粹的符号。

要让计算机学会区分第二个单引号，我们需要用到转义字符`\`。

```
print('I\'m not your brother.')
```
这样一来，计算机在读到第二个单引号前的`\`时，就会恍然大悟：哦！原来这个单引号和第一个单引号不是一起的呀，那我就把它当做要打印的内容里的一部分吧！

其实，除了`\`，在前面我们已经介绍过一种让计算机能区分符号功能的方法了。

# 变量和赋值

我们先来看看下面这行代码：

```
name='巴基'
```
这就是一个常见的“给变量赋值”的动作。在这里，`name`就是一个变量，这行代码的意思，就是把`巴基`这两个个字【赋值】给了`name`这个【变量】。

这就好比，为了让巴基在日后能在计算机中更方便地找到他的名字，我们帮他把名字收纳进了一个小盒子里，并在这个盒子上贴了个叫`name`的标签。

可是，我们为什么要对信息进行“收纳”呢？

因为，每台计算机都要存储和操作成千上万的数据，这就等同于我们家里囤了成千上万的东西一般，不加以整理的话，根本找不到想要的东西在哪。

为了让家看上去更整洁，我们可以把不同的东西放进不同的盒子里，并且做好标记。这样家里不仅会整整齐齐的，也会更方便我们自己取用东西。

同理，计算机也是这么做的，它会把成千上万的数据放在不同的“盒子”里，这样才方便它存储和操作数据。

这个“盒子”在Python中被称为变量，你可以在这个“盒子”里放任何你想放的内容。

而这个往盒子里装东西的过程，在代码的世界被称为【赋值】。

**变量就像是一个带标签的盒子，需要你把数据放进去。**

巴基是美国队长史蒂夫青梅竹马的好朋友，二战时期他与美国队长多次出生入死对抗红骷髅，在某次执行任务期间，巴基掉下万丈深渊阵亡。不为人知的是，九头蛇最先发现并解冻了他，且还在他的断臂处安装上了机械手臂，并将其洗脑训练成了效忠九头蛇的特工“冬日战士”。

于是，“巴基”这两个字对于他来说成为了一个陌生的字眼，他再也不记得原本的名字了。

这时，装巴基名字的“盒子”就被重新赋值变成了：

```
name='冬日战士'
```

```
name='巴基'

print(name)
```

````python
# 直接运行代码，看一下打印函数的最终输出结果是什么
name='巴基'
name='冬日战士'
print(name)
````

你的结果是什么，是不是`冬日战士`？可这又是为什么呢？我们对`name`这个变量第一次的赋值不是`'巴基'`吗？为什么打印的结果是第二次的赋值呢？

这就涉及到了变量的特点：变量之所以称为变量，是因为它保存的数据是可以随意变化的。

刚才我们讲，我们可以将变量当作一个盒子，你可以在这个盒子里放任何内容，但盒子都有它的最大容量，不能无限制地塞东西。

偏偏在代码世界里，盒子的容量又特别小，仅为1。所以当你放进新的东西时，旧的东西就会被替代掉。

被多次**赋值**后，变量中储存的结果永远是**最新**一次的赋值。

在我们上面的案例中，第一行代码：`name='巴基'`表达的意思是：把`巴基`这个名字放到`name`这个变量“盒子”中。

由于，计算机是一行行自上而下执行代码的，所以当运行到第二行代码`name='冬日战士'`时，变量`name`里存储的`'巴基'`就被替换成了`'冬日战士'`。

所以，当运行到第三行`print(name)`时，我们打印出的结果，就自然而然的是`冬日战士`了。

就这样，巴基坠入深谷后被九头蛇组织的人擒获，将他救活并控制，阿尼姆·佐拉博士在巴基的大脑中植入了精神控制器，清除了他从前的记忆，剥夺了他的自我。其实，不仅是巴基，名字对于我们每个人来讲都具有非同寻常的意义，它从出生起就跟随我们，帮我们形成独立的个体。

而给变量取名字也是如此，我们要注意变量的命名规范，才能让我们在纷繁复杂的代码世界，更为方便地使用它。

## 变量的命名与规范

所以，在变量与赋值中，你要注意的第一个规则就是：你需要学会给变量取一个合适的、标准的名字。

对于变量的命名，我们可以遵循以下规范：

- 只能是一个词。

- 只能包含字母、数字和下划线。

- 不能以数字开头。

- 尽量描述包含的数据内容。

比如说，如果信息是姓名，那么变量可以命名为`name`；如果信息是数字，那么变量名则应该叫做`number`。

第二点要注意的规则是：在代码世界中，赋值符号=不是左边等于右边的意思。仅仅表示赋值动作：把右边的内容放到了左边的盒子中。

代表左右两边相等的符号，是比较运算符==，虽然他俩长得像，但代表的却是完全不同的意思，可千万不要搞混了。
```
a = 2 赋值
a = = 2 等于
```
